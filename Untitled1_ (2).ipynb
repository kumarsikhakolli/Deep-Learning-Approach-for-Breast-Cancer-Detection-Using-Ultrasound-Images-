```python
import os, shutil, random
random.seed(42)

BASE_DIR = "/content/drive/MyDrive/Dataset_BUSI_with_GT"
OUT_DIR  = "/content/drive/MyDrive/BUSI_SPLIT"

classes = ["benign", "malignant"]   # ignore normal for cancer classification

splits = {
    "train": 0.7,
    "val": 0.15,
    "test": 0.15
}
```

```python
for split in splits:
    for cls in classes:
        os.makedirs(f"{OUT_DIR}/{split}/{cls}", exist_ok=True)
```

```python
for cls in classes:
    img_dir = f"{BASE_DIR}/{cls}"

    images = [f for f in os.listdir(img_dir) if f.endswith((".png", ".jpg"))]
    random.shuffle(images)

    n = len(images)
    train_end = int(n * splits["train"])
    val_end   = int(n * (splits["train"] + splits["val"]))

    split_map = {
        "train": images[:train_end],
        "val":   images[train_end:val_end],
        "test":  images[val_end:]
    }

    for split, files in split_map.items():
        for f in files:
            shutil.copy(
                f"{img_dir}/{f}",
                f"{OUT_DIR}/{split}/{cls}/{f}"
            )
```

```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
from torchvision import transforms

train_tfms = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.5],[0.5])
])

val_tfms = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5],[0.5])
])
```

```python
from torchvision import datasets
from torch.utils.data import DataLoader

DATA_DIR = "/content/drive/MyDrive/BUSI_SPLIT"

train_data = datasets.ImageFolder(f"{DATA_DIR}/train", transform=train_tfms)
val_data   = datasets.ImageFolder(f"{DATA_DIR}/val", transform=val_tfms)

train_loader = DataLoader(train_data, batch_size=16, shuffle=True)
val_loader   = DataLoader(val_data, batch_size=16, shuffle=False)

num_classes = len(train_data.classes)
print(train_data.classes)
```

```python
import torch.nn as nn

class CBAM(nn.Module):
    def __init__(self, channels, reduction=16):
        super().__init__()
        self.avg = nn.AdaptiveAvgPool2d(1)
        self.max = nn.AdaptiveMaxPool2d(1)

        self.fc = nn.Sequential(
            nn.Conv2d(channels, channels//reduction, 1, bias=False),
            nn.ReLU(),
            nn.Conv2d(channels//reduction, channels, 1, bias=False)
        )

        self.spatial = nn.Conv2d(2, 1, kernel_size=7, padding=3)

    def forward(self, x):
        ca = torch.sigmoid(self.fc(self.avg(x)) + self.fc(self.max(x)))
        x = x * ca

        sa = torch.sigmoid(
            self.spatial(
                torch.cat([x.mean(1,True), x.max(1,True)[0]], dim=1)
            )
        )
        return x * sa
```

```python
import timm

class EfficientNet_CBAM(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.backbone = timm.create_model(
            "efficientnet_b0",
            pretrained=True,
            features_only=True
        )
        self.cbam = CBAM(320)
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Linear(320, num_classes)

    def forward(self, x):
        x = self.backbone(x)[-1]
        x = self.cbam(x)
        x = self.pool(x).flatten(1)
        return self.fc(x)
```

```python
model = EfficientNet_CBAM(num_classes).to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
def train_epoch():
    model.train()
    loss_sum = 0
    for x,y in train_loader:
        x,y = x.to(device), y.to(device)

        optimizer.zero_grad()
        out = model(x)
        loss = criterion(out,y)
        loss.backward()
        optimizer.step()

        loss_sum += loss.item()
    return loss_sum / len(train_loader)
```

```python
def val_epoch():
    model.eval()
    correct,total = 0,0
    with torch.no_grad():
        for x,y in val_loader:
            x,y = x.to(device), y.to(device)
            pred = model(x).argmax(1)
            correct += (pred==y).sum().item()
            total += y.size(0)
    return correct/total
```

```python
for epoch in range(15):
    loss = train_epoch()
    acc  = val_epoch()
    print(f"Epoch {epoch+1} | Loss {loss:.4f} | Val Acc {acc:.4f}")
```

```python
test_data = datasets.ImageFolder(
    f"{DATA_DIR}/test",
    transform=val_tfms
)

test_loader = DataLoader(test_data, batch_size=16, shuffle=False)
print("Test samples:", len(test_data))
```

```python
def test_epoch():
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for x,y in test_loader:
            x,y = x.to(device), y.to(device)
            pred = model(x).argmax(1)
            correct += (pred == y).sum().item()
            total += y.size(0)
    return correct / total

test_acc = test_epoch()

print("Test Accuracy:", test_acc)

```

```python
from PIL import Image
import torch
import torch.nn.functional as F

def predict_image(img_path):
    img = Image.open(img_path).convert("RGB")
    img_t = val_tfms(img).unsqueeze(0).to(device)

    model.eval()
    with torch.no_grad():
        logits = model(img_t)
        probs = F.softmax(logits, dim=1)

        pred_class = probs.argmax(1).item()
        confidence = probs[0][pred_class].item()

    return train_data.classes[pred_class], confidence
```
